```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("Customer churn.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df["TotalCharges"]= df["TotalCharges"].replace(" ","0") 
df["TotalCharges"]= df["TotalCharges"].astype("float")
```

```python
df.head()
```

```python
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plot

ax = sns.countplot(x='Churn', data=df)


ax.bar_label(ax.containers[0])

plt.show()

```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(aspect="equal"))

gb = df.groupby("Churn").agg({'Churn': "count"})


colors = ['#00008B', '#FFFFFF']


wedges, texts, autotexts = ax.pie(gb['Churn'], labels=gb.index, autopct="%1.2f%%", colors=colors, startangle=90, wedgeprops={'edgecolor': 'black'}, explode=(0.1, 0))


plt.title("Percentage of Churned Customers")

for text in texts + autotexts:
    text.set_fontsize(12)
    text.set_color('black')

plt.show()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")

plt.figure(figsize=(6, 4))
ax = sns.countplot(x="gender", data=df, hue="Churn", palette=["#00008B", "#FFFFFF"])

for container in ax.containers:
    ax.bar_label(container, label_type='edge', color='black')  # Change the color to black for better visibility

plt.title("Gender Distribution by Churn")
 
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_grouped = df.groupby(['SeniorCitizen', 'Churn']).size().unstack().fillna(0)
df_grouped = df_grouped.apply(lambda x: x / x.sum() * 100, axis=1)

df_grouped.plot(kind='bar', stacked=True, color=["#00008B", "#FFFFFF"], figsize=(6, 4))

for n, x in enumerate([*df_grouped.index.values]):
    for (proportion, y_loc) in zip(df_grouped.loc[x], df_grouped.loc[x].cumsum()):
        plt.text(x=n, y=y_loc - (proportion / 2), s=f'{proportion:.1f}%', color="black", ha="center", va="center")

plt.title("Percentage of Churn by Senior Citizen Status")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 15))

axes = axes.flatten()

for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, hue='Churn', palette=["#00008B", "#FFFFFF"], ax=axes[i])
    axes[i].set_title(f'Count of Customers by {col}')
    axes[i].bar_label(axes[i].containers[0], label_type='edge', color='black')
    if len(axes[i].containers) > 1:
        axes[i].bar_label(axes[i].containers[1], label_type='edge', color='black')


plt.tight_layout()


plt.show()
```
